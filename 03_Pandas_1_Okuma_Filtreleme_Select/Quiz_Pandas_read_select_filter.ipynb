## 1) Sentetik veriyi olustur ve CSV olarak kaydet


```python
# Bu hücreyi calistir: veri oluşacak ve CSV kaydedilecek.
np.random.seed(21)

n = 40
siparis_id = np.arange(1001, 1001+n)

sehir = np.random.choice(["Istanbul","Ankara","Izmir","Bursa"], size=n, p=[0.45,0.25,0.2,0.1])
segment = np.random.choice(["Student","Employee","Visitor"], size=n, p=[0.5,0.35,0.15])
urun = np.random.choice(["Coffee","Dessert","Sandwich","Juice"], size=n, p=[0.35,0.25,0.25,0.15])

birim_fiyat = []
for u in urun:
    if u == "Coffee":
        birim_fiyat.append(np.random.randint(40, 90))
    elif u == "Dessert":
        birim_fiyat.append(np.random.randint(70, 140))
    elif u == "Sandwich":
        birim_fiyat.append(np.random.randint(90, 180))
    else:
        birim_fiyat.append(np.random.randint(60, 130))
birim_fiyat = np.array(birim_fiyat)

adet = np.random.randint(1, 5, size=n)
indirim_orani = np.round(np.random.choice([0.00,0.05,0.10,0.15], size=n, p=[0.55,0.2,0.2,0.05]), 2)

teslim = np.random.choice(["Delivered","Cancelled"], size=n, p=[0.92,0.08])
puan = np.random.choice([1,2,3,4,5], size=n, p=[0.05,0.10,0.25,0.35,0.25])

# Son 90 gun icinde rastgele tarih
base = pd.Timestamp("2026-01-29")
gun_ofset = np.random.randint(0, 90, size=n)
siparis_tarihi = (base - pd.to_timedelta(gun_ofset, unit="D")).strftime("%Y-%m-%d")

df = pd.DataFrame({
    "siparis_id": siparis_id,
    "sehir": sehir,
    "segment": segment,
    "urun": urun,
    "birim_fiyat": birim_fiyat,
    "adet": adet,
    "indirim_orani": indirim_orani,
    "siparis_tarihi": siparis_tarihi,
    "teslim_durumu": teslim,
    "puan": puan
})

csv_adi = "ybs_kafe_siparisleri.csv"
df.to_csv(csv_adi, index=False)

print("CSV oluştu:", csv_adi)
```
