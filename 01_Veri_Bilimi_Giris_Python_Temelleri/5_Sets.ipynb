# Python Temelleri — Set (Küme)

Bu notebookta şunları öğreneceğiz:
- Set nedir? `type()` ile kontrol
- Set’in en önemli özelliği: **tekrar yok (unique)** ve **sırasız**
- Eleman ekleme/silme: `add()`, `remove()`, `discard()`
- Üyelik kontrolü: `in`
- Küme işlemleri: birleşim, kesişim, fark


## 1) Set nedir? (tekrar eden elemanlar otomatik silinir)


```python
etiketler = {"python", "veri", "python", "ybs"}
print(etiketler)
```

## 2) Tip kontrolü: `type()`


```python
etiketler = {"python", "veri", "ybs"}
print(type(etiketler))  # set
```

## 3) Set oluşturma (boş set)


Dikkat: `{}` boş sözlüktür. Boş set için `set()` kullanılır.


```python
bos_set = set()
bos_sozluk = {}

print(type(bos_set))    # set
print(type(bos_sozluk)) # dict
```

## 4) `len()` ile eleman sayısı


```python
dersler = {"Python", "SQL", "Pandas"}
print(len(dersler))
```

## 5) Eleman ekleme: `add()`


```python
dersler = {"Python", "SQL"}
dersler.add("Pandas")
print(dersler)
```

## 6) Eleman silme: `remove()` vs `discard()`


- `remove()` eleman yoksa hata verir
- `discard()` eleman yoksa hata vermez


```python
dersler = {"Python", "SQL", "Pandas"}

dersler.remove("SQL")
print(dersler)

# discard hata vermez
dersler.discard("Excel")
print(dersler)
```

## 7) Üyelik kontrolü: `in`


```python
dersler = {"Python", "SQL", "Pandas"}

print("Python" in dersler)
print("Excel" in dersler)
```

## 8) Set gezme (for)


Set sırasızdır, ekrana geliş sırası farklı olabilir.


```python
dersler = {"Python", "SQL", "Pandas"}

for ders in dersler:
    print("Ders:", ders)
```

## 9) Küme işlemleri (Union / Intersection / Difference)


```python
a = {"python", "sql", "pandas"}
b = {"sql", "excel", "power"}

print("Birleşim (|):", a | b)
print("Kesişim (&):", a & b)
print("Fark (a - b):", a - b)
print("Fark (b - a):", b - a)
```

## 10) Listeden tekrarları temizleme


```python
musteriler = ["Ali", "Ayşe", "Ali", "Mehmet", "Ayşe"]

benzersiz_musteriler = set(musteriler)
print(benzersiz_musteriler)

# tekrar listeye çevirelim (sıra garanti değil)
musteri_listesi = list(benzersiz_musteriler)
print(musteri_listesi)
```

## 11) Mini alıştırma — Ortak dersleri bulma


İki öğrencinin aldığı dersler set olarak verilmiş. Ortak dersleri yazdırın.


```python
ogrenci1 = {"Python", "SQL", "Pandas"}
ogrenci2 = {"SQL", "Excel", "Python"}

ortak = ogrenci1 & ogrenci2
print("Ortak dersler:", ortak)
```
