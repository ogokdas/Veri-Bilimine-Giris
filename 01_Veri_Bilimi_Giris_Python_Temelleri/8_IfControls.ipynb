# Python Temelleri — If Kontrolleri

Bu notebookta şunları öğreneceğiz:
- `if`, `elif`, `else` yapısı
- Karşılaştırmalar (`==`, `!=`, `>`, `<`, `>=`, `<=`)
- Mantıksal operatörler (`and`, `or`, `not`)
- İç içe if (nested if)
- Pratik mini örnekler


## 1) En basit `if`


```python
yas = 20

if yas >= 18:
    print("Giriş serbest")
```

## 2) `if` + `else`


```python
yas = 16

if yas >= 18:
    print("Giriş serbest")
else:
    print("Giriş yasak")
```

## 3) `if` + `elif` + `else`


```python
not_ortalamasi = 78

if not_ortalamasi >= 85:
    print("Harf notu: AA")
elif not_ortalamasi >= 70:
    print("Harf notu: BB")
else:
    print("Harf notu: CC")
```

## 4) `and` ile iki şart birlikte


```python
yas = 22
gelir = 15000

if yas >= 18 and gelir >= 10000:
    print("Kredi başvurusu uygun")
else:
    print("Kredi başvurusu uygun değil")
```

## 5) `or` ile en az bir şart


```python
ogrenci_mi = True
personel_mi = False

if ogrenci_mi or personel_mi:
    print("İndirim uygulanır")
else:
    print("İndirim yok")
```

## 6) `not` ile tersini alma


```python
aktif_mi = False

if not aktif_mi:
    print("Hesap pasif")
else:
    print("Hesap aktif")
```

## 7) İç içe if (nested if)


```python
yas = 20
kimlik_var_mi = True

if yas >= 18:
    if kimlik_var_mi:
        print("İçeri girebilir")
    else:
        print("Kimlik gerekli")
else:
    print("Yaş tutmuyor")
```

## 8) Kısa if


```python
puan = 55
sonuc = "Geçti" if puan >= 50 else "Kaldı"
print(sonuc)
```

## 9) Mini örnek — Kargo bedava mı?


1000 TL ve üzeri alışverişte kargo bedava.


```python
sepet_tutari = 1200

if sepet_tutari >= 1000:
    print("Kargo bedava ✅")
else:
    print("Kargo ücreti var ❌")
```

## 10) Mini örnek — Giriş doğrulama


```python
kullanici_adi = "omer"
sifre = "1234"

if kullanici_adi == "omer" and sifre == "1234":
    print("Giriş başarılı")
else:
    print("Hatalı kullanıcı adı veya şifre")
```
