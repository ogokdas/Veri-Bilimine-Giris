# Python Temelleri — Variables & Numbers

Bu notebookta şunları öğreneceğiz:
- Değişken (variable) nedir, nasıl atanır?
- Sayı tipleri: `int`, `float`
- `type()` ile tip kontrolü
- Temel işlemler: `+ - * / // % **`
- Tip dönüşümü: `int()`, `float()`, `round()`
- f-string ile yazdırma


## 1) Değişken atama


```python
yas = 21
ucret = 19.99
isim = "Ayşe"
is_active = True

print(yas, ucret, isim, is_active)
```

## 2) Tip kontrolü: `type()`


```python
yas = 21
ucret = 19.99

print(type(yas))     # int
print(type(ucret))   # float
```

## 3) `int` ve `float` (Numbers)


```python
a = 10      # int
b = 3.5     # float

print(a, type(a))
print(b, type(b))
```

## 4) Temel aritmetik işlemler


```python
x = 12
y = 5

print("Toplama:", x + y)
print("Çıkarma:", x - y)
print("Çarpma:", x * y)
print("Bölme (/):", x / y)
print("Tam bölme (//):", x // y)
print("Kalan (%):", x % y)
print("Üs (**):", x ** y)
```

## 5) İşlem önceliği (parantez farkı)


```python
print(2 + 3 * 4)     # 14
print((2 + 3) * 4)   # 20
```

## 6) Tip dönüşümü (Casting)


```python
s1 = "10"
s2 = "3.14"

n1 = int(s1)
n2 = float(s2)

print(n1, type(n1))
print(n2, type(n2))
```

## 7) Yuvarlama: `round()`


```python
not_ort = 3.45678
print(round(not_ort, 2))
```

## 8) f-string ile yazdırma


```python
isim = "Ali"
yas = 22
not_ort = 3.45678

print(f"{isim} {yas} yaşında.")
print(f"Not ortalaması: {not_ort:.2f}")
```

## 9) Mini alıştırma (KDV dahil fiyat)


```python
ucret = 850
kdv_oran = 0.20

toplam = ucret * (1 + kdv_oran)
print(f"KDV dahil fiyat: {toplam:.2f} TL")
```

## 10) Mini alıştırma (Kutulama)


```python
items = 53
box_capacity = 6

full_boxes = items // box_capacity
kalan = items % box_capacity

print("Tam dolan kutu:", full_boxes)
print("Artan ürün:", kalan)
```
