# Python Temelleri — String (Metin)

Bu notebookta şunları öğreneceğiz:
- String nedir? `type()` ile kontrol
- Birleştirme (concat) ve tekrar
- Uzunluk: `len()`
- Karakterlere erişim: indeksleme
- Parça alma: dilimleme (slicing)
- Sık kullanılan metotlar: `lower()`, `upper()`, `strip()`, `replace()`, `split()`
- f-string ile yazdırma


## 1) String nedir?


```python
ad = "Ayşe"
sehir = "İstanbul"
mesaj = "Merhaba"

print(ad)
print(sehir)
print(mesaj)
```

## 2) Tip kontrolü: `type()`


```python
ad = "Ayşe"
yas = 21

print(type(ad))   # str
print(type(yas))  # int
```

## 3) String birleştirme ve tekrar


```python
ad = "Ali"
soyad = "Yılmaz"

tam_isim = ad + " " + soyad
print(tam_isim)

print("Ha" * 3)  # HaHaHa
```

## 4) Uzunluk: `len()`


```python
urun_adi = "Laptop"
print(len(urun_adi))
```

## 5) İndeksleme (karakterlere erişim)


Python’da indeksler **0**’dan başlar. Negatif indeks de kullanılabilir.


```python
kelime = "Python"

print(kelime[0])   # P
print(kelime[1])   # y
print(kelime[-1])  # n (son karakter)
```

## 6) Dilimleme (Slicing)


```python
kelime = "VeriBilimi"

print(kelime[0:4])   # Veri
print(kelime[4:])    # Bilimi
print(kelime[:4])    # Veri
```

## 7) Sık kullanılan string metotları


```python
metin = "  Merhaba Dünya  "

print(metin.lower())
print(metin.upper())
print(metin.strip())
```

```python
cumle = "veri bilimi çok eğlenceli"
print(cumle.replace("eğlenceli", "keyifli"))
```

### 7.1) `split()` ile parçalama


```python
cumle = "Python, Pandas, NumPy"
parcalar = cumle.split(", ")
print(parcalar)
```

## 8) Metin içinde arama: `in` ve `find()`


```python
mail = "omergokdas@gmail.com"

print("@" in mail)      # True / False
print(mail.find("@"))   # varsa index, yoksa -1
```

## 9) f-string ile yazdırma


```python
ad = "Zeynep"
urun = "Kulaklık"
fiyat = 799.9

print(f"{ad} adlı müşteri {urun} aldı. Tutar: {fiyat:.2f} TL")
```

## 10) Mini alıştırmalar


### Alıştırma 1 — E-posta kullanıcı adı çıkarma


```python
eposta = "kullanici@mail.com"
kullanici_adi = eposta.split("@")[0]
print(kullanici_adi)
```

### Alıştırma 2 — Baş harfleri alma


```python
ad = "Mehmet"
soyad = "Demir"

bas_harfler = ad[0] + soyad[0]
print(bas_harfler)
```

### Alıştırma 3 — Temizleme (strip)


```python
girilen = "   veri bilimi   "
print(girilen.strip())
```
