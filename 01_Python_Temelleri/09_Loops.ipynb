# Python Temelleri — Döngüler (for / while)

Bu notebookta şunları öğreneceğiz:
- `for` döngüsü (liste, range)
- `while` döngüsü
- `break` ve `continue`
- Basit örnekler ve mini alıştırmalar


## 1) `for` döngüsü (liste üzerinde)


```python
urunler = ["Laptop", "Mouse", "Klavye"]

for urun in urunler:
    print("Ürün:", urun)
```

## 2) `for` + `range()`


```python
for i in range(5):
    print(i)  # 0,1,2,3,4
```

## 3) `range(start, stop, step)`


```python
for i in range(2, 11, 2):
    print(i)  # 2,4,6,8,10
```

## 4) Toplam alma örneği


```python
notlar = [70, 85, 90]

toplam = 0
for not_degeri in notlar:
    toplam += not_degeri

ortalama = toplam / len(notlar)
print("Toplam:", toplam)
print("Ortalama:", ortalama)
```

## 5) index + değer


```python
urunler = ["Laptop", "Mouse", "Klavye"]
index = 0
for urun in urunler:
    print(index, "->", urun)
    index = index+1
```

## 6) `while` döngüsü (şart sağlandığı sürece)


```python
sayac = 1

while sayac <= 5:
    print("Sayaç:", sayac)
    sayac += 1
```

## 7) `break` (döngüyü bitirir)


```python
for sayi in range(1, 10):
    if sayi == 5:
        break
    print(sayi)
```

## 8) `continue` (o turu atlar)


```python
for sayi in range(1, 6):
    if sayi == 3:
        continue
    print(sayi)
```

## 9) Mini örnek — Şifre kontrol (while)


Doğru şifre girilene kadar sorar (input kısmı yorumlu).


```python
sifre = ""
while sifre != "1234":
  sifre = input("Şifre girin: ")
print("Giriş başarılı")
```

## 10) Mini alıştırma 1 — 1'den 10'a kadar çift sayılar


```python
for sayi in range(1, 11):
    if sayi % 2 == 0:
        print(sayi)
```

## 11) Mini alıştırma 2 — Sepet toplamı


Bir sepet fiyat listesinin toplamını bulun.


```python
fiyatlar = [120, 55, 300, 80]

toplam = 0
for fiyat in fiyatlar:
    toplam += fiyat

print("Sepet toplamı:", toplam)
```

## 12) Mini alıştırma 3 — En büyük değeri bulma


```python
puanlar = [70, 85, 90, 60]

en_buyuk = puanlar[0]
for p in puanlar:
    if p > en_buyuk:
        en_buyuk = p

print("En büyük puan:", en_buyuk)
```
