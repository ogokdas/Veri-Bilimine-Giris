# Python Temelleri — Functions (Fonksiyonlar)

Bu notebookta şunları öğreneceğiz:

- `def` ile fonksiyon yazma
- Parametre (girdi) alma
- `return` ile çıktı döndürme
- Varsayılan (default) parametre

## 1) Fonksiyon nedir?


Fonksiyon: Tekrar eden işleri **tek yerde** yazıp gerektiğinde çağırmamızı sağlar.


## 2) En basit fonksiyon


```python
def selam_ver():
    print("Merhaba!")

selam_ver()
selam_ver()
```

## 3) Parametre alan fonksiyon


```python
def karsila(isim):
    print(f"Hoş geldin {isim}!")

karsila("Ayşe")
karsila("Mehmet")
```

## 4) `return` ile değer döndürme


```python
def toplam(a, b):
    sonuc = a + b
    return sonuc

t = toplam(10, 5)
print(t)
```

## 5) Varsayılan (default) parametre


```python
def kdv_hesapla(fiyat, kdv_orani=0.20):
    return fiyat * (1 + kdv_orani)

print(kdv_hesapla(1000))        # %20 varsayılan
print(kdv_hesapla(1000, 0.10))  # %10
```

## 6) Fonksiyon içinde if kullanımı


```python
def indirim_var_mi(sepet_tutari):
    if sepet_tutari >= 1000:
        return True
    else:
        return False

print(indirim_var_mi(850))
print(indirim_var_mi(1200))
```

## 7) Birden fazla değer döndürme


Python'da fonksiyon birden fazla değer döndürebilir (tuple olarak).


```python
def sepet_ozet(fiyatlar):
    toplam = sum(fiyatlar)
    adet = len(fiyatlar)
    ortalama = toplam / adet
    return toplam, adet, ortalama

fiyatlar = [120, 55, 300, 80]
toplam, adet, ortalama = sepet_ozet(fiyatlar)

print("Toplam:", toplam)
print("Ürün sayısı:", adet)
print("Ortalama:", ortalama)
```

## 8) Mini alıştırmalar


### Alıştırma 1 — Ortalama not


3 not alan bir fonksiyon yazın ve ortalamayı döndürün.


```python
def ortalama_not(n1, n2, n3):
    return (n1 + n2 + n3) / 3

print(ortalama_not(70, 85, 90))
```

### Alıştırma 2 — Kullanıcı adı çıkarma


Bir e-posta alıp kullanıcı adını döndüren fonksiyon yazın.


```python
def kullanici_adi_bul(eposta):
    return eposta.split("@")[0]

print(kullanici_adi_bul("ogokdas1@gmail.com"))
```

### Alıştırma 3 — En yüksek fiyat


Fiyat listesini alıp en yüksek fiyatı döndüren fonksiyon yazın.


```python
def en_yuksek_fiyat(fiyatlar):
    return max(fiyatlar)

print(en_yuksek_fiyat([120, 55, 300, 80]))
```
