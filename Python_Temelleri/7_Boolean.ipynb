# Python Temelleri — Boolean (Mantıksal)

Bu notebookta şunları öğreneceğiz:
- Boolean nedir? (`True` / `False`) ve `type()`
- Karşılaştırma operatörleri: `== != > < >= <=`
- Mantıksal operatörler: `and`, `or`, `not`
- `in` ile kontrol
- True/False döndüren pratik örnekler
- Basit `if` örneği


## 1) Boolean nedir?


```python
aktif_mi = True
ogrenci_mi = False

print(aktif_mi)
print(ogrenci_mi)
```

## 2) Tip kontrolü: `type()`


```python
aktif_mi = True
print(type(aktif_mi))  # bool
```

## 3) Karşılaştırma operatörleri


```python
yas = 21

print(yas == 21)   # eşit mi?
print(yas != 18)   # eşit değil mi?
print(yas > 18)    # büyük mü?
print(yas < 30)    # küçük mü?
print(yas >= 21)   # büyük-eşit mi?
print(yas <= 20)   # küçük-eşit mi?
```

## 4) Mantıksal operatörler: `and`, `or`, `not`


```python
yas = 21
gelir = 120000

# and: ikisi de True olmalı
print(yas > 18 and gelir > 100000)

# or: en az biri True ise True
print(yas < 18 or gelir > 100000)

# not: tersine çevirir
aktif_mi = True
print(not aktif_mi)
```

## 5) `in` ile kontrol (string / list içinde var mı?)


```python
mail = "omergokdas@gmail.com"
print("@" in mail)

urunler = ["Laptop", "Mouse", "Klavye"]
print("Mouse" in urunler)
print("Telefon" in urunler)
```

## 6) True/False döndüren basit örnekler


```python
sifre = "1234"
print(len(sifre) >= 8)  # şifre en az 8 karakter mi?

stok = 0
print(stok > 0)         # stok var mı?
```

## 7) Basit `if` örneği


```python
yas = 17

if yas >= 18:
    print("Giriş serbest")
else:
    print("Giriş yasak")
```

## 8) Mini alıştırma — Kampanya kontrolü


1000 TL ve üzeri alışverişte kargo bedava olsun.


```python
sepet_tutari = 850
kargo_bedava_mi = sepet_tutari >= 1000

print("Kargo bedava mı?", kargo_bedava_mi)
```
