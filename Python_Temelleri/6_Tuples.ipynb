# Python Temelleri — Tuple

Bu notebookta şunları öğreneceğiz:
- Tuple nedir? `type()` ile kontrol
- Tuple vs List (en büyük fark: **değiştirilemez / immutable**)
- Elemanlara erişim (index) ve dilimleme (slicing)
- `len()` ile uzunluk
- Tuple açma (unpacking) ve çoklu atama
- `count()` ve `index()`


## 1) Tuple nedir?


```python
sehirler = ("İstanbul", "Ankara", "İzmir")
puanlar = (70, 85, 90)

print(sehirler)
print(puanlar)
```

## 2) Tip kontrolü: `type()`


```python
sehirler = ("İstanbul", "Ankara", "İzmir")
print(type(sehirler))  # tuple
```

## 3) Tek elemanlı tuple (dikkat!)


Tek elemanlı tuple için sonda virgül olmalı: `('Python',)`


```python
tek = ("Python",)
liste_gibi = ("Python")

print(type(tek))        # tuple
print(type(liste_gibi)) # str
```

## 4) Tuple elemanlarına erişim (index)


```python
sehirler = ("İstanbul", "Ankara", "İzmir")

print(sehirler[0])   # İstanbul
print(sehirler[-1])  # İzmir
```

## 5) Dilimleme (slicing)


```python
sehirler = ("İstanbul", "Ankara", "İzmir", "Bursa", "Antalya")

print(sehirler[0:3])  # ilk 3 şehir
print(sehirler[2:])   # 2. indexten sonrası
```

## 6) Uzunluk: `len()`


```python
sehirler = ("İstanbul", "Ankara", "İzmir")
print(len(sehirler))
```

## 7) Tuple değiştirilemez (immutable)


Tuple içindeki elemanları sonradan değiştiremezsiniz.


```python
sehirler = ("İstanbul", "Ankara", "İzmir")

# Aşağı satır hata verir (yorumdan çıkarıp deneyebilirsiniz)
# sehirler[0] = "Bursa"

print(sehirler)
```

## 8) Tuple açma (unpacking)


```python
urun = ("Laptop", 35000, 12)  # (ad, fiyat, stok)

urun_adi, fiyat, stok = urun
print(urun_adi)
print(fiyat)
print(stok)
```

## 9) Çoklu atama ile takas (swap)


```python
a = 5
b = 12

a, b = b, a
print("a:", a, "b:", b)
```

## 10) `count()` ve `index()`


```python
veriler = (10, 20, 10, 30, 10)

print(veriler.count(10))  # kaç tane 10 var?
print(veriler.index(30))  # 30 hangi indexte?
```
